### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

total_tienda1 = sum(tienda['Precio'])
total_tienda2 = sum(tienda2['Precio'])
total_tienda3 = sum(tienda3['Precio'])
total_tienda4 = sum(tienda4['Precio'])

totales = [
    sum(tienda['Precio']/1000000),
    sum(tienda2['Precio']/1000000),
    sum(tienda3['Precio']/1000000),
    sum(tienda4['Precio']/1000000)
]
#convertido a millones para legibilidad

```

```python
print(f'El ingreso total de la primera tienda es {total_tienda1}')
print(f'El ingreso total de la segunda tienda es {total_tienda2}')
print(f'El ingreso total de la tercera tienda es {total_tienda3}')
print(f'El ingreso total de la cuarta tienda es {total_tienda4}')
```

# 2. Ventas por categoría

```python
# En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.


conteo_tienda1 = tienda["Categoría del Producto"].value_counts()
conteo_tienda2 = tienda2["Categoría del Producto"].value_counts()
conteo_tienda3 = tienda3["Categoría del Producto"].value_counts()
conteo_tienda4 = tienda4["Categoría del Producto"].value_counts()



```

```python
print("Conteo de productos por categoría para Tienda 1:")
print(conteo_tienda1)
print("\nConteo de productos por categoría para Tienda 2:")
print(conteo_tienda2)
print("\nConteo de productos por categoría para Tienda 3:")
print(conteo_tienda3)
print("\nConteo de productos por categoría para Tienda 4:")
print(conteo_tienda4)





```

# 3. Calificación promedio de la tienda


```python
#En este paso, debes calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.

calificacion_promedio_tienda1= (sum(tienda['Calificación']))/(len(tienda['Calificación']))
calificacion_promedio_tienda2= (sum(tienda2['Calificación']))/(len(tienda['Calificación']))
calificacion_promedio_tienda3= (sum(tienda3['Calificación']))/(len(tienda['Calificación']))
calificacion_promedio_tienda4= (sum(tienda4['Calificación']))/(len(tienda['Calificación']))


calificaciones = [
    calificacion_promedio_tienda1,
    calificacion_promedio_tienda2,
    calificacion_promedio_tienda3,
    calificacion_promedio_tienda4
]

```

```python
print(f'La calificación promedio de la primera tienda es: {calificacion_promedio_tienda1}')
print(f'La calificación promedio de la segunda tienda es: {calificacion_promedio_tienda2}')
print(f'La calificación promedio de la tercera tienda es: {calificacion_promedio_tienda3}')
print(f'La calificación promedio de la cuarta tienda es: {calificacion_promedio_tienda4}')


```

# 4. Productos más y menos vendidos

```python
#En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

producto_mas_vendido_tienda1 = tienda['Producto'].value_counts().idxmax()
producto_menos_vendido_tienda1 = tienda['Producto'].value_counts().idxmin()

producto_mas_vendido_tienda2 = tienda2['Producto'].value_counts().idxmax()
producto_menos_vendido_tienda2 = tienda2['Producto'].value_counts().idxmin()

producto_mas_vendido_tienda3 = tienda3['Producto'].value_counts().idxmax()
producto_menos_vendido_tienda3 = tienda3['Producto'].value_counts().idxmin()

producto_mas_vendido_tienda4 = tienda4['Producto'].value_counts().idxmax()
producto_menos_vendido_tienda4 = tienda4['Producto'].value_counts().idxmin()
```

```python
print(f'El producto más vendido de la primera tienda es: {producto_mas_vendido_tienda1}')
print(f'El producto menos vendido de la primera tienda es: {producto_menos_vendido_tienda1}')
print(f'\nEl producto más vendido de la segunda tienda es: {producto_mas_vendido_tienda2}')
print(f'El producto menos vendido de la segunda tienda es: {producto_menos_vendido_tienda2}')
print(f'\nEl producto más vendido de la tercera tienda es: {producto_mas_vendido_tienda3}')
print(f'El producto menos vendido de la tercera tienda es: {producto_menos_vendido_tienda3}')
print(f'\nEl producto más vendido de la cuarta tienda es: {producto_mas_vendido_tienda4}')
print(f'El producto menos vendido de la cuarta tienda es: {producto_menos_vendido_tienda4}')




```

# 5. Envío promedio por tienda

```python
envio_promedio_tienda1= (sum(tienda['Costo de envío']))/(len(tienda['Costo de envío']))
envio_promedio_tienda2= (sum(tienda2['Costo de envío']))/(len(tienda2['Costo de envío']))
envio_promedio_tienda3= (sum(tienda3['Costo de envío']))/(len(tienda3['Costo de envío']))
envio_promedio_tienda4= (sum(tienda4['Costo de envío']))/(len(tienda4['Costo de envío']))

```

```python
print(f'El envío promedio de la primera tienda es: {envio_promedio_tienda1}')
print(f'El envío promedio de la segunda tienda es: {envio_promedio_tienda2}')
print(f'El envío promedio de la tercera tienda es: {envio_promedio_tienda3}')
print(f'El envío promedio de la cuarta tienda es: {envio_promedio_tienda4}')

envios = [
    envio_promedio_tienda1/1000,
    envio_promedio_tienda2/1000,
    envio_promedio_tienda3/1000,
    envio_promedio_tienda4/1000
]
```

```python
import matplotlib.pyplot as plt




tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']


plt.bar(tiendas, totales, color='skyblue')
plt.title("Ventas totales por tienda")
plt.ylabel("Total en millones $")
plt.xlabel("Tiendas")
plt.ylim(1020, 1160)

for i, val in enumerate(totales):
    plt.text(i, val + 0.01, f"{val:.2f}", ha='center')


plt.show()
```

```python
import numpy as np


df = pd.DataFrame({
    "Tienda 1": conteo_tienda1,
    "Tienda 2": conteo_tienda2,
    "Tienda 3": conteo_tienda3,
    "Tienda 4": conteo_tienda4
}).fillna(0)

categorias = df.index
tiendas = df.columns
x = np.arange(len(categorias))  # posiciones en X
width = 0.2  # ancho de cada barra

fig, ax = plt.subplots(figsize=(14, 7))

# Colores (puedes cambiar aquí si lo prefieres)
colores = ["#1f77b4", "#2ca02c", "#ff7f0e", "#d62728"]

# Dibujar barras para cada tienda
for i, tienda in enumerate(tiendas):
    bars = ax.bar(x + (i - 1.5) * width, df[tienda], width, label=tienda, color=colores[i])
    # Agregar números encima de las barras
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f"{int(height)}",
                    xy=(bar.get_x() + bar.get_width()/2, height),
                    xytext=(0, 3),  # desplazamiento
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=8)

# Personalización
ax.set_ylabel("Cantidad de Productos")
ax.set_xlabel("Categorías")
ax.set_title("Conteo de Productos por Categoría y Tienda")
ax.set_xticks(x)
ax.set_xticklabels(categorias, rotation=45, ha='right')
ax.legend()
plt.tight_layout()
plt.show()
```

```python

plt.barh(tiendas, calificaciones, color='green')
plt.title("Calificación promedio por tienda")
plt.xlabel("Calificación escala 1 a 5")
plt.ylabel("Tiendas")
plt.xlim(3.5, 4.1)  # horizontal → ajusta eje X en vez de Y

for i, val in enumerate(calificaciones):
    plt.text(val + 0.01, i, f"{val:.2f}", va='center')  # cambia coordenadas

plt.show()
```

```python


# Datos
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
mas_vendidos = [
    producto_mas_vendido_tienda1,
    producto_mas_vendido_tienda2,
    producto_mas_vendido_tienda3,
    producto_mas_vendido_tienda4
]
menos_vendidos = [
    producto_menos_vendido_tienda1,
    producto_menos_vendido_tienda2,
    producto_menos_vendido_tienda3,
    producto_menos_vendido_tienda4
]

df_productos = pd.DataFrame({
    "Más Vendido": mas_vendidos,
    "Menos Vendido": menos_vendidos
}, index=tiendas)

# Gráfico horizontal
fig, ax = plt.subplots(figsize=(10, 6))

y = range(len(tiendas))
width = 0.4

bars1 = ax.barh(
    [i + width/2 for i in y],
    [1]*len(tiendas),
    height=width,
    label="Más Vendido",
    color="green"
)
bars2 = ax.barh(
    [i - width/2 for i in y],
    [1]*len(tiendas),
    height=width,
    label="Menos Vendido",
    color="red"
)

# Anotar los nombres de los productos
for i, bar in enumerate(bars1):
    ax.annotate(
        df_productos.iloc[i, 0],
        xy=(1.02, bar.get_y() + bar.get_height()/2),
        xytext=(5, 0),
        textcoords="offset points",
        va="center",
        fontsize=9
    )
for i, bar in enumerate(bars2):
    ax.annotate(
        df_productos.iloc[i, 1],
        xy=(1.02, bar.get_y() + bar.get_height()/2),
        xytext=(5, 0),
        textcoords="offset points",
        va="center",
        fontsize=9
    )

ax.set_yticks(y)
ax.set_yticklabels(tiendas)
ax.set_xlim(0, 1.5)
ax.set_xticks([])
ax.set_xlabel("")
ax.set_title("Productos Más y Menos Vendidos por Tienda", pad=20)
ax.legend(loc="lower right")

plt.tight_layout()
plt.show()
```
